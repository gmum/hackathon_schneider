```python
from datetime import datetime
import pandas as pd
import json

#loading the processed dataframe
df4 = pd.read_csv('../data/df_processed.csv')
```

```python
#ploting to check surplus values
df_surplus = df4.filter(like='surplus')
df_surplus.reset_index(inplace=True)
df_total_surplus = pd.DataFrame({'Total_surplus': df_surplus.sum(axis=1)})

import matplotlib.pyplot as plt

plt.figure(figsize=(20, 2))
plt.hist(df_total_surplus['Total_surplus'], bins=50, color='blue', alpha=0.7)
plt.title('Surplus Distribution')
plt.xlabel('Surplus')
plt.ylabel('Frequency')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming df_surplus contains the relevant data

# Set the style of seaborn
sns.set(style="whitegrid")

# Create boxplots for each column
plt.figure(figsize=(20, 3))
sns.boxplot(data=df_surplus, palette="Set3")

# Customize the plot
plt.title('Distribution of Surplus Values')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better visibility
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming df_surplus contains the relevant data

# Set the style of seaborn
sns.set(style="whitegrid")

# Calculate the total for each column
df_surplus['Total'] = df_surplus.sum(axis=1)

# Create a line plot
plt.figure(figsize=(20, 3))
sns.lineplot(x='StartTime', y='Total', data=df_surplus)

# Customize the plot
plt.title('Total Surplus Over Time')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better visibility
plt.show()
```
